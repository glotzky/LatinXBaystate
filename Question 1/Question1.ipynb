```python
import numpy as np
import pandas as pd

#How has support for Trump shifted across the LatinX population from 2016 to 2020? Was there a significant shift?
```

```python

pres2016 = pd.read_csv("pres_demographic_voting_2016.csv")

pres2020 = pd.read_csv("pres_demographic_voting_2020.csv")

gov2014 = pd.read_csv("gov_demographic_voting_2014.csv")

gov2018 = pd.read_csv("gov_demographic_voting_2018.csv")
```

```python
#cleaning for presidential election

#take our comma values
pres2020 = pres2020.replace(",", "", regex = True)

pres2016 = pres2016.replace(",", "", regex = True)

pres2020["Joseph R. Biden, Jr."] = pd.to_numeric(pres2020["Joseph R. Biden, Jr."])

pres2020["Donald J. Trump"] = pd.to_numeric(pres2020["Donald J. Trump"])

pres2020["Total Votes Cast"] = pd.to_numeric(pres2020["Total Votes Cast"])

pres2020["Estimate Mexican Population"] = pd.to_numeric(pres2020["Estimate Mexican Population"])

pres2020["Estimate Puerto Rican Population"] = pd.to_numeric(pres2020["Estimate Puerto Rican Population"])

pres2020["Estimate Cuban Population"] = pd.to_numeric(pres2020["Estimate Cuban Population"])

pres2020["Estimate Other Hispanic or Latino"] = pd.to_numeric(pres2020["Estimate Other Hispanic or Latino"])

pres2016["Hillary Clinton"] = pd.to_numeric(pres2016["Hillary Clinton"])

pres2016["Donald J. Trump"] = pd.to_numeric(pres2016["Donald J. Trump"])

pres2016["Total Votes Cast"] = pd.to_numeric(pres2016["Total Votes Cast"])

pres2016["Estimate Mexican Population"] = pd.to_numeric(pres2016["Estimate Mexican Population"])

pres2016["Estimate Puerto Rican Population"] = pd.to_numeric(pres2016["Estimate Puerto Rican Population"])

pres2016["Estimate Cuban Population"] = pd.to_numeric(pres2016["Estimate Cuban Population"])

pres2016["Estimate Other Hispanic or Latino"] = pd.to_numeric(pres2016["Estimate Other Hispanic or Latino"])
```

```python
pres2020.head()
```

```python
#cleaning for governors election

gov2014 = gov2014.replace(",", "", regex = True)

gov2018 = gov2018.replace(",", "", regex = True)

gov2018["Gonzalez and Palfrey (D)"] = pd.to_numeric(gov2018["Gonzalez and Palfrey (D)"])

gov2018["Baker and Polito (R)"] = pd.to_numeric(gov2018["Baker and Polito (R)"])

gov2018["Total Votes Cast"] = pd.to_numeric(gov2018["Total Votes Cast"])

gov2018["Estimate Mexican Population"] = pd.to_numeric(gov2018["Estimate Mexican Population"])

gov2018["Estimate Puerto Rican Population"] = pd.to_numeric(gov2018["Estimate Puerto Rican Population"])

gov2018["Estimate Cuban Population"] = pd.to_numeric(gov2018["Estimate Cuban Population"])

gov2018["Estimate Other Hispanic or Latino"] = pd.to_numeric(gov2018["Estimate Other Hispanic or Latino"])

gov2014["Coakley and Kerrigan (D)"] = pd.to_numeric(gov2014["Coakley and Kerrigan (D)"])

gov2014["Baker and Polito (R)"] = pd.to_numeric(gov2014["Baker and Polito (R)"])

gov2014["Total Votes Cast"] = pd.to_numeric(gov2014["Total Votes Cast"])

gov2014["Estimate Mexican Population"] = pd.to_numeric(gov2014["Estimate Mexican Population"])

gov2014["Estimate Puerto Rican Population"] = pd.to_numeric(gov2014["Estimate Puerto Rican Population"])

gov2014["Estimate Cuban Population"] = pd.to_numeric(gov2014["Estimate Cuban Population"])

gov2014["Estimate Other Hispanic or Latino"] = pd.to_numeric(gov2014["Estimate Other Hispanic or Latino"])
```

```python
pres2016['Total LatinX Pop'] = pres2016['Estimate Mexican Population'] + pres2016['Estimate Puerto Rican Population'] + pres2016['Estimate Cuban Population'] + pres2016['Estimate Other Hispanic or Latino']                   

pres2020['Total LatinX Pop'] = pres2020['Estimate Mexican Population'] + pres2020['Estimate Puerto Rican Population'] + pres2020['Estimate Cuban Population'] + pres2020['Estimate Other Hispanic or Latino']    


gov2014['Total LatinX Pop'] = gov2014['Estimate Mexican Population'] + gov2014['Estimate Puerto Rican Population'] + gov2014['Estimate Cuban Population'] + gov2014['Estimate Other Hispanic or Latino']  

gov2018['Total LatinX Pop'] = gov2018['Estimate Mexican Population'] + gov2018['Estimate Puerto Rican Population'] + gov2018['Estimate Cuban Population'] + gov2018['Estimate Other Hispanic or Latino']  
```

```python
pres2016['LatinX Percent'] = pres2016['Total LatinX Pop'] / pres2016['Estimate Total Population']

pres2020['LatinX Percent'] = pres2020['Total LatinX Pop'] / pres2020['Estimate Total Population']


gov2014['LatinX Percent'] = gov2014['Total LatinX Pop'] / gov2014['Estimate Total Population']

gov2018['LatinX Percent'] = gov2018['Total LatinX Pop'] / gov2018['Estimate Total Population']
```

```python
pres2016['Republican Percent'] = pres2016['Donald J. Trump'] / pres2016['Total Votes Cast']

pres2020['Republican Percent'] = pres2020['Donald J. Trump'] / pres2020['Total Votes Cast']


pres2016['Democratic Percent'] = pres2016['Hillary Clinton'] / pres2016['Total Votes Cast']

pres2020['Democratic Percent'] = pres2020['Joseph R. Biden, Jr.'] / pres2020['Total Votes Cast']
```

```python
gov2014['Republican Percent'] = gov2014['Baker and Polito (R)'] / gov2014['Total Votes Cast']

gov2018['Republican Percent'] = gov2018['Baker and Polito (R)'] / gov2018['Total Votes Cast']


gov2014['Democratic Percent'] = gov2014['Coakley and Kerrigan (D)'] / gov2014['Total Votes Cast']

gov2018['Democratic Percent'] = gov2018['Gonzalez and Palfrey (D)'] / gov2018['Total Votes Cast']
```

```python
gov2014.head()
```

```python
pres2016 = pres2016.add_suffix('_2016')

pres2020 = pres2020.add_suffix('_2020')

gov2014 = gov2014.add_suffix('_2014')

gov2018 = gov2018.add_suffix('_2018')
```

```python
pres2016.rename(columns={"Tracts_2016": "Tracts"}, inplace = True)

pres2020.rename(columns={"Tracts_2020": "Tracts"}, inplace = True)

gov2014.rename(columns={"Tracts_2014": "Tracts"}, inplace = True)

gov2018.rename(columns={"Tracts_2018": "Tracts"}, inplace = True)
```

```python
merged_pres = pd.merge(pres2016, pres2020, on="Tracts") 

merged_gov = pd.merge(gov2014, gov2018, on="Tracts") 
```

```python
merged_gov.head()
```

```python
#percent change 

merged_pres['Republican Change'] = (merged_pres['Republican Percent_2020'] - merged_pres['Republican Percent_2016']) 

merged_pres['Democratic Change'] = (merged_pres['Democratic Percent_2020'] - merged_pres['Democratic Percent_2016'])

merged_pres['LatinX Pop Change'] = (merged_pres['LatinX Percent_2020'] - merged_pres['LatinX Percent_2016']) 


merged_gov['Republican Change'] = (merged_gov['Republican Percent_2018'] - merged_gov['Republican Percent_2014']) 

merged_gov['Democratic Change'] = (merged_gov['Democratic Percent_2018'] - merged_gov['Democratic Percent_2014']) 

merged_gov['LatinX Pop Change'] = (merged_gov['LatinX Percent_2018'] - merged_gov['LatinX Percent_2014'])  
```

```python
merged_pres_processed = merged_pres[['Republican Change', 'Democratic Change', 'LatinX Pop Change']]

merged_gov_processed = merged_gov[['Republican Change', 'Democratic Change', 'LatinX Pop Change']]
```

```python
merged_pres_processed.head()
```

```python
merged_gov_processed.head()
```

```python
#regression for Rebublican Presidential 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

X = merged_pres_processed.iloc[:, 2].values.reshape(-1, 1)  
Y_rep = merged_pres_processed.iloc[:, 0].values.reshape(-1, 1)  
regr_rep = LinearRegression()
regr_rep.fit(X, Y_rep)
Y_pred_rep = regr_rep.predict(X)

plt.scatter(X, Y_rep, color = 'red', label = 'Rebublican Support')
plt.plot(X, Y_pred_rep, color='lightsalmon', label = 'Republican Support Trend')
plt.xlabel('LatinX Pop Percentage Point Change')
plt.ylabel('Republican Votes Percentage Point Change')
plt.title('Change in total LatinX Population vs change in Rebublican Election Results for Presidental Election (2016-2020)')
plt.legend(loc="upper right")
plt.show()

```

```python
#regression for Democratic Presidential 

X = merged_pres_processed.iloc[:, 2].values.reshape(-1, 1)  
Y_dem = merged_pres_processed.iloc[:, 1].values.reshape(-1, 1)  
regr_dem = LinearRegression()
regr_dem.fit(X, Y_dem)
Y_pred_dem = regr_dem.predict(X)

plt.scatter(X, Y_dem, color='blue', label = 'Democratic Support')
plt.plot(X, Y_pred_dem, color='cyan', label = 'Democratic Support Trend')
plt.xlabel('LatinX Pop Percentage Point Change')
plt.ylabel('Republican Votes Percentage Point Change')
plt.title('Change in total LatinX Population vs change in Democratic Election Results for Presidental Election (2016-2020)')
plt.legend(loc="upper right")
plt.show()
```

```python
#regression for Rebublican and Democratic Presidential

plt.scatter(X, Y_rep, color = 'red', label = 'Rebublican Support')
plt.scatter(X, Y_dem, color='blue', label = 'Democratic Support')


plt.plot(X, Y_pred_rep, color='lightsalmon', label = 'Republican Support Trend')
plt.plot(X, Y_pred_dem, color='cyan', label = 'Democratic Support Trend')
plt.xlabel('LatinX Pop Percentage Point Change')
plt.ylabel('Rep and Dem Percentage Point Change')
plt.title('Change in total LatinX Population vs change in Democratic Election Results for Presidental Election (2016-2020)')
plt.legend(loc="upper right")
plt.show()
```
